```python
dataset = dataset.loc[:,['battery_power','mobile_wt','px_height','px_width','ram']]
dataset
```

```python
from pandas import test
from scipy.sparse import data
from sklearn.preprocessing import MinMaxScaler

sc = MinMaxScaler(feature_range=(0,1))
test_data = sc.fit_transform(dataset)
print(test_data)
```

```python
import pandas as pd
from sklearn.cluster import KMeans

max_iterasi = 10
cluster_i = []
cluster_val = [] 
cluster = []

clustering = KMeans(n_clusters=3, init="random", n_init=1)


for i in range(max_iterasi):
    fit =  clustering.fit(test_data)
    clusters = clustering.predict(test_data)
    sse = clustering.inertia_
    print("\nHasil clustering iterasi ke ",i+1,":\n", clusters)
    print("\nSSE iterasi ke ",i+1,":\n", sse)
    print("------------------------------------------------------")
    cluster_i.append(clusters)
    cluster_val.append(sse)

min_sse = min(cluster_val)

index_min = cluster_val.index(min_sse)

cluster = cluster_i[index_min]


```
